### Exploratory Data Analysis
Exploratory Data Analysis refers to the critical process of performing initial investigations on data so as to discover patterns, to spot anomalies, to test hypothesis and to check assumptions with the help of summary statistics and graphical representations. EDA analyse and identifies general patterns in data. These patterns includes outliers and features of the data that might be unexpected. EDA is an important first step in any data analysis. Understanding where outliers occur and how variables are related can help one design statistical analyses that yield meaningful results.

Analyzing nearest_warehouse



```python
df['Nearest warehouse'].unique()
```

Checking quantities of near_warehouse

```python
warehouses =df['Nearest warehouse'].value_counts()
print(warehouses)
```

### ***Visualizing data using bar chart***

```python
warehouses.plot(kind='bar', rot=1, color=['r','g','b'], figsize=(10,5))
plt.title("Quantity of the warehouses\n", size=15)
plt.ylabel("Counts");
```

### ***Visualizing data using pie chart***

```python
plt.pie(warehouses,
        labels=['Thomson',"Nickolson",'Bakers','thomson','nickolson'],
        autopct="%0.1f%%",
        explode=[0,.01,0.001,0.0001,0.1],
        startangle=60,radius=1)
plt.title("Quantity of warehouses", size=15)
plt.show()
```

It seems that Thomson is nearest warehouse accessible to mostly customer

Analyzing how many customers are happy with service



```python
df['Is happy customer'].unique()
satisfaction = df['Is happy customer'].value_counts()
print(satisfaction)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
satisfaction.plot(kind='bar', color=['b','r'], rot=1)
plt.xticks(range(2), labels=['Satisfy','Not Satifty'])
plt.ylabel('Counts')
plt.title("Customer satisfaction Vs Not satisfaction\n", size=15)


plt.subplot(122)
plt.pie(satisfaction, labels=['Satisfy','Not Satifty'], autopct="%.1f%%",explode=[0,0.1],shadow=True)
plt.title("Percentage of Customer satisfaction Vs Not satisfaction\n", size=15)

plt.show()
```

71.7% customers are satisfy with services
